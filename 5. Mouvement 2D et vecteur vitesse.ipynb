# Fiche n°5 : Mouvement 2D et vecteur vitesse

On étudie le mouvement d’une balle de golf à partir d’une chronophotographie dans le but est de tracer les vecteurs vitesse au cours du temps. Les mêmes outils présentés dans les activités précédentes sont réexploités. Les positions relevées toutes les 0,066 s sont rassemblées dans un tableau.

**Capacité numérique** :
* représenter les positions successives d’un système modélisé par un point lors d’une évolution unidimensionnelle ou bidimensionnelle à l’aide d’un langage de programmation.
* représenter des vecteurs vitesse d’un système modélisé par un point lors d’un mouvement à l’aide d’un langage de programmation.

## 1. Représentation de la trajectoire

Le tableau des temps successifs, puis ceux des abscisses et des ordonnées sont créés. Pour représenter la trajectoire, il faut faire tracer l’ordonnée en fonction de l’abscisse.

```python
import numpy as np
import matplotlib.pyplot as plt

t=0.066*np.arange(11) # ici le nombre de mesures est N=11
x=np.array([0.01,0.25,0.57,0.91,1.22,1.54,1.87,2.16,2.49,2.81,3.15])
y=np.array([0.01549,0.3404,0.6812,1.010,1.297,1.559,1.768,1.95,2.08,2.158,2.193])
plt.plot(x,y,'ro',label="y=f(x)")
plt.xlabel("xreelle")
plt.ylabel("yreelle")
plt.grid()
plt.legend()
plt.title("Trajectoire de la balle")
plt.show()
```

## 2. Représentation des vecteurs déplacement 

Comme précédemment, la boucle **for** permet de faire défiler les positions successives (termes de yreelle) et le tracé des vecteurs s’effectue avec la fonction **plt.arrow**. Comme précédemment, il faut prendre garde à redimensionner la liste, pour qu’elle contienne un élément de moins (car on ne peut pas calculer x[i+1] et y[i+1] pour la dernière valeur). La commande **len**(t) indique le nombre de valeurs dans la liste t.


```python
n=np.arange(len(t)-1)
for i in n :
    plt.arrow(x[i],y[i],x[i+1]-x[i],y[i+1]-y[i],head_width=0.05) # on fait tracer une série de vecteurs
                                                                 # de position initiale x[i],y[i], de coordonnées x[i+1]-x[i] et y[i+1]-y[i]
                                                                 # head_width est la largeur de la tête
plt.xlim(0,4)
plt.ylim(0,3)
plt.grid()

plt.title("Vecteurs déplacement")
plt.show()
```

**Conclusion** : le résultat est plus lisible. 

## 3. Représentation des vecteurs vitesse

Pour les vecteurs vitesse, plusieurs définitions sont envisageables. La plus classique consiste à calculer le déplacement entre les points i et i+2, à diviser par l’intervalle de temps qui sépare ces deux positions et à affecter cette vitesse au point i+1. Ainsi, il faut à nouveau redimensionner la liste des i comme cela est proposé dans le script suivant :

```python
m=np.arange(len(t)-2) # comme on travaille dans la suite sur des intervalles i à i+2, il faut couper la boucle à N-2

for i in m :
    # on fait tracer une série de vecteurs de position initiale x[i+1],y[i+1], de coordonnées 0.1(x[i+2]-x[i])/(t[i+2]-t[i])
    # et idem selon uy. 
    # Ces longueurs correspondent au déplacement de i à i+2 rapporté à l'intervalle de temps correspondant.
    # on multiplie par 0.1 pour que la figure soit lisible.    
    plt.arrow(x[i+1],y[i+1],0.1*(x[i+2]-x[i])/(t[i+2]-t[i]),0.1*(y[i+2]-y[i])/(t[i+2]-t[i]),head_width=0.05) 

plt.xlim(0,4)
plt.ylim(0,3)
plt.grid()
```

**Conclusion** : On notera que si le résultat est plus lisible, il faut tout de même réfléchir à l’échelle de représentation du vecteur vitesse sur le graphe - c’est la raison du facteur 0.1 qui a été placé devant les vecteurs.
