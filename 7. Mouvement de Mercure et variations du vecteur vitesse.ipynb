# Fiche n°7 : Mouvement de Mercure et variations du vecteur vitesse

L’étude porte sur le mouvement de Mercure. Le but est de tracer les variations du vecteur vitesse au cours du temps. Les positions de Mercure ont été obtenues sur le site des éphémérides pour 11 dates démarrant le 1/1/2017 et espacées de 8 jours. Le principe est exactement le même que dans les fiches précédentes. 


**Capacités numériques** : 
* représenter les positions successives d’un système modélisé par un point lors d’une évolution unidimensionnelle ou bidimensionnelle à l’aide d’un langage de programmation.
* représenter des vecteurs vitesse d’un système modélisé par un point lors d’un mouvement à l’aide d’un langage de programmation.
* représenter des vecteurs variation de vitesse d’un système modélisé par un point lors d’une évolution unidimensionnelle ou bidimensionnelle à l’aide d’un langage de programmation.


## 1. Représentation de la trajectoire


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=8*np.arange(11)
Xmes=np.array([-0.2,2.9,5.4,6.4,5,1.2,-3.4,-6.4,-6.9,-5.7,-3.3])
Ymes=np.array([-8.2,-7.3,-4.9,-1.2,2.9,5.3,4.7,1.5,-2.3,-5.5,-7.5])
x=Xmes*0.46/8.2
y=Ymes*0.46/8.2
plt.plot(x,y,'ro')
plt.xlabel("xlue")
plt.ylabel("ylue")
plt.grid()
plt.axis('equal')
plt.title("Position de Mercure en fonction du temps")
plt.show()
```

## 2. Représentation des vecteurs vitesse  

```python
m=np.arange(len(t)-2)
for i in m :
    plt.arrow(x[i+1],y[i+1],5*(x[i+2]-x[i])/(t[i+2]-t[i]),5*(y[i+2]-y[i])/(t[i+2]-t[i]),head_width=0.05)
plt.plot(x,y,'bo')
plt.axis('equal')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.title("Vecteurs déplacement")
plt.grid()
plt.show()
```

## 3. Construction des vecteurs variation de vitesse 

```python
vxl=[]
m=np.arange(len(t)-2)
for i in m :
    vx=(x[i+2]-x[i])/(t[i+2]-t[i])
    vxl.append(vx)

vyl=[]
m=np.arange(len(t)-2)
for i in m :
    vy=(y[i+2]-y[i])/(t[i+2]-t[i])
    vyl.append(vy)
```

```python
axl=[]
m=np.arange(len(t)-4)
for i in m :
    ax=(vxl[i+2]-vxl[i])/(t[i+2]-t[i])
    axl.append(ax)

ayl=[]
m=np.arange(len(t)-4)
for i in m :
    ay=(vyl[i+2]-vyl[i])/(t[i+2]-t[i])
    ayl.append(ay)

t3=t[:-4]
```

```python
m=np.arange(1,len(t)-4)
for i in m :
    plt.arrow(x[i+2],y[i+2],200*axl[i],200*ayl[i],head_width=0.02)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.axis('equal')
plt.grid()
plt.plot(x,y,'bo')
```
